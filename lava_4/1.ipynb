### Пробуємо побудувати прості графіки

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10, 10)
y = 10 + x/4
print(f"""
Згенеровано значень: {x.size}
Значення по x: {x}
Значення по y: {y}
""")
plt.plot(x,y)
```

```python
x = np.linspace(-10, 10, num=20)
y = x**2 + 2*x + 10
print(f"""
Згенеровано значень: {x.size}
Значення по x: {x}
Значення по y: {y}
""")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.grid()
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-7, 5, num=20)
y_a = 30 + x/4
y_b = x**2 + 2*x + 10

b = y_a - y_b
print(b)

plt.plot(x, y_a, "--", label="30 + x/4")
plt.plot(x, y_b, "-", label="x**2 + 2*x + 10")
plt.legend()
plt.grid()
plt.show()
```
